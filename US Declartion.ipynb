```python
f = open('US_Declaration.pdf','rb')

#list of every paage's text.
#the index wioll corspond to the page number
pdf_text = []

pdf_reader = PyPDF2.PdfReader(f)

for p in range(len(pdf_reader.pages)):

    page = pdf_reader.pages[p]
    pdf_text.append(page.extract_text())

f.close()
```

```python
pdf_text
```

```python
len(pdf_text)
```

```python
for page in pdf_text:
    print(page)
    print('\n')
    print('\n')
    print('\n')
    print('\n')
    print('\n')
    print('\n')
```
